### Pascal VOC Format

Create XML files in the Pascal VOC Format for the SMALL TRAIN+TEST

```python
import pandas as pd
import numpy as np
import ast
import os
from pascal_voc_writer import Writer
```

Read text file with annotations

```python
ann = pd.read_csv('annotated.txt',header=None)
ann
```

Obtain train and test file names

```python
train = [f for f in os.listdir("train/") if f.endswith('.jpg')]
test = [f for f in os.listdir("test/") if f.endswith('.jpg')]
```

Create XML for train

```python
ann_train = ann[ann[0].isin(train)]

for r in range(len(ann_train)):
    # Writer(path, width, height)
    writer = Writer("train/" + ann_train.iloc[r, 0], 1501, 944)

    # ::addObject(name, xmin, ymin, xmax, ymax)
    writer.addObject(ann_train.iloc[r, 5], ann_train.iloc[r, 1], ann_train.iloc[r, 2], ann_train.iloc[r, 3], ann_train.iloc[r, 4])

    # ::save(path)
    writer.save("train/" + ann_train.iloc[r, 0][:-4] + ".xml")
```

Create XML for test

```python
ann_test = ann[ann[0].isin(test)]

for r in range(len(ann_test)):
    # Writer(path, width, height)
    writer = Writer("test/" + ann_test.iloc[r, 0], 1501, 944)

    # ::addObject(name, xmin, ymin, xmax, ymax)
    writer.addObject(ann_test.iloc[r, 5], ann_test.iloc[r, 1], ann_test.iloc[r, 2], ann_test.iloc[r, 3], ann_test.iloc[r, 4])

    # ::save(path)
    writer.save("test/" + ann_test.iloc[r, 0][:-4] + ".xml")
```
